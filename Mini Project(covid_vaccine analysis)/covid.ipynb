To check whether dataset having any null values....

```python
data.isnull().sum()
```

Yes the dataset do have a null values thus replacing it by a mean(average)......

For first Dose Administered

```python
avg_firstdose = data["First Dose Administered"].astype("float").mean(axis = 0)
print("Average of First Dose:", avg_firstdose)
```

```python
# Replacing First Dose Administered
```

```python
data["First Dose Administered"].fillna(value = avg_firstdose, inplace=True)
data
```

```python
For Second Dose Administered
```

```python
avg_seconddose = data["Second Dose Administered"].astype("float").mean(axis = 0)
print("Average of Second Dose:", avg_seconddose)
```

# Replacing Second Dose Administered

```python
data["Second Dose Administered"].fillna(value = avg_seconddose, inplace = True)
data
```

```python
Now we have treated the null values now let us find 
Administration of first and second dose statewise
```

Number of persons state wise vaccinated for first dose in India

```python
first_dose = data.groupby('State')[['First Dose Administered']].sum()
first_dose
```

Number of persons state wise vaccinated for second dose in India

```python
first_dose = data.groupby('State')[['Second Dose Administered']].sum()
first_dose
```

Number of Males vaccinated

```python
male = data["Male(Individuals Vaccinated)"].sum()
print("The total number of male individuals vaccinated are", int(male))
```

Number of Females vaccinated

```python
female = data["Female(Individuals Vaccinated)"].sum()
print("The total number of female individuals vaccinated are", int(female))
```

Grouping the data by State and aggregating the sum of Male/Female (Individuals Vaccinated)

```python
gender_df = data.groupby("State")[["Male (Doses Administered)", "Female (Doses Administered)"]].sum()
```

```python
ax = gender_df.plot(kind='bar', stacked=True, figsize=(8,8))
plt.xlabel("State/UT")
plt.ylabel("Number of Individuals Vaccinated")
plt.title("Number of Males and Females Vaccinated State wise")

plt.show()
```

# Grouping the data by State and aggregating the sum of First Dose Administered and Second Dose Administered

```python
dose_df = data.groupby("State")["First Dose Administered", "Second Dose Administered"].sum()

# Creating a grouped bar plot
ax = dose_df.plot(kind='bar', figsize=(8,8))

# Adding labels and title
plt.xlabel("State/UT")
plt.ylabel("Number of Doses Administered")
plt.title("Number of People Vaccinated for First and Second Dose State wise")

# Displaying the plot
plt.show()
```

Plotting a piechart  of vaccination of first dose statewise in India 

```python
first_dose = data.groupby('State')['First Dose Administered'].max()

# Create a pie chart to show the distribution of first dose vaccinations by state
plt.pie(first_dose, labels=first_dose.index, autopct='%1.1f%%')
plt.title("State-wise Distribution of First Dose Vaccinations")
plt.show()
```

```python
second_dose=data.groupby('State')['Second Dose Administered'].max()
plt.pie(second_dose, labels=second_dose.index, autopct='%1.1f%%')
plt.title("State-wise Distribution of Second Dose Vaccinations")
plt.show()
```
